```python
# FULLY OBJECT ORIENTATED
import os
import json 
import urllib.request

cityFile = "city.list.json"


'''

1. CREATE A CLASS
2. THAT IS CONTAINER
3. DEFINE FUNCTION TO RETURN DATA 
4. ASSIGN FUNCTION TO VAR
5. WORK THROUGH VAR/FUNCTION - GETTING EACH VALUE, AND PASSING INTO A TEMP ARRAY THAT IS ASSIGNED THE CLASS/CONTAINER
'''



# CREATE A CLASS / CONTAINER
class createCityArray:
    def __init__(self, coord,country,cityId,name): # initialize   
        self.coord = coord
        self.country = country
        self.cityId = cityId # can't use id
        self.name = name




def getCityData():
    json_data=open(cityFile).read()
    data = json.loads(json_data)
    return data

print("Fetching city Data from file ...")

cityData = getCityData()
tempArray = []
cityArray = []

print("Data retreived, filling city objects")
if "error" not in cityData:
    for item in cityData:
        coord = str(item.get("coord"))
        country = str(item.get("country"))
        cityId = str(item.get("id"))
        name = str(item.get("name"))
        
        # THIS IS CREATION OF OBJECT, CONSTRUCTING (POPULATION)
        tempArray = createCityArray(coord, country, cityId,name)
        print('hi')
    cityArray.append(tempArray)
    
print(cityArray)
```

```python
import APIconnect
import urllib, json
import urllib.request

# OPEN FILE
filename = 'weatherReport.json' 
f = open(filename,'w')
selectedCity = ["3333229","2648110", "3333225"] # SELECT CITIES FOR REPORTING
weatherReport = "" # Empty var to store responses
weatherArray = []

allData = ""
for city in range(0, len(selectedCity)):
    
    
    # Define the full URL with connecton key to allow transactions, city changes each iteration
    URL = "http://api.openweathermap.org/data/2.5/weather?id=" + selectedCity[city] + "&APPID=" + APIconnect.URLkey
    with urllib.request.urlopen(URL) as url:
        data = json.loads(url.read().decode())
        print(data)
    weatherArray.append(data)
    
    
    
jsonweatherArray = json.dumps(weatherArray)

f.write(str(jsonweatherArray))
f.close() 
    
    

```
